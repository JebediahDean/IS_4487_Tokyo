```python
grp_category = ss_clean.groupby(["order_year_month", "product_line"])["quantity"].sum()
grp_category
```

```python
import matplotlib.pyplot as plt

# Extract data for plotting
months = grp_category.index.get_level_values(0).unique()
product_lines = grp_category.index.get_level_values(1).unique()

# Create a figure and axes
plt.figure(figsize=(12, 8))

# Iterate over product lines and plot bubbles
for product_line in product_lines:
    x = []
    y = []
    sizes = []
    for month in months:
        try:
            quantity = grp_category.loc[(month, product_line)]
            x.append(month)
            y.append(product_line)
            sizes.append(quantity * 0.5)  # Adjust scaling factor as needed
        except KeyError:
            pass  # Handle cases where a product line has no sales in a specific month
    plt.scatter(x, y, s=sizes, alpha=0.7, label=product_line)

# Add labels and title
plt.xlabel('Order Year-Month')
plt.ylabel('Product Line')
plt.title('Total Quantity Sold by Product Line and Month (Bubble Chart)')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Display legend and grid
plt.legend(title='Product Line')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
