# Importing Libraries and Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/jefftwebb/is_4487_base/main/Assignments/DataSets/adviseinvest.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Parsing Data

```python
# Removing Negatives
df_cleaning = df[df["income"] > 0]
```

```python
# Creating Categorical Variables
df_cleaning["job"] = df_cleaning["job"].astype("category")
df_cleaning["product"] = df_cleaning["product"].astype("category")
df_cleaning["chk_acct"] = df_cleaning["chk_acct"].astype("category")
df_cleaning["sav_acct"] = df_cleaning["sav_acct"].astype("category")
```

```python
# Null Values
df_cleaning.isnull().sum()
```

```python
df_cleaning = df_cleaning.dropna()
```

```python
# Removing Mistakes in the Data
```

```python
plt.boxplot(df["num_accts"], labels=["Number of Accounts"])
plt.show()
```

```python
df_cleaning = df_cleaning[df_cleaning["num_accts"] < 10 ]
```

```python
plt.boxplot(df_cleaning["num_accts"], labels=["Number of Accounts"])
```

```python
df_cleaning.head()
```

```python
df_cleaning["answered"]=df_cleaning["answered"].replace(1,"yes")
df_cleaning["answered"]=df_cleaning["answered"].replace(0,"no")
df_cleaning["answered"]=df_cleaning["answered"].astype("category")
```

```python
df_cleaning.info()
```

```python
df_clean = df_cleaning.copy()
```

# Data Analysis

## Problem 1

```python
df_clean.head(2)
```

```python
plt.boxplot([df_clean[df_clean["answered"]=="yes"]['income'],
            df_clean[df_clean["answered"]=="no"]['income']],
            labels=["yes","no"])

plt.title("Income by Answered")
plt.ylabel("Income")
plt.xlabel("Answered")

plt.show()
```

```python
sns.boxplot(x="answered", y="income", data=df_clean)
plt.show()
```

The graph shows that if a customer has more money, between $60000 or more, they are more lickly to not answer the phone call. However, if a customer has less than $4500 in income the customer could either answer or not.  

## Question 2

```python
# Create a barplot of the counts or proportions of customers who answered, comparing those with and without mobile phones.
```

```python
df_clean.head(2)
```

```python
df_clean["mobile"]=df_clean["mobile"].replace(1,"Answered")
df_clean["mobile"]=df_clean["mobile"].replace(0,"Unresponsive")
df_clean["mobile"]=df_clean["mobile"].astype("category")
```

```python
df_clean.head(2)
```

```python
grp_mobile = df_clean.groupby(["mobile","answered"])["answered"].count().unstack()
grp_mobile
```

```python
(df_clean.groupby(["mobile","answered"])["answered"]
 .count()
 .unstack()
 .plot(kind="bar", xlabel="Responsiveness", ylabel="Counts", title="Counts of Responsibeness for Sales Calls"))
```

This shows that the count of people who have a mobile phone answer slightly more often than people who don't. Additionally, people who don't have a phobile phone are unresponsive the slightly more often than people who don't.
